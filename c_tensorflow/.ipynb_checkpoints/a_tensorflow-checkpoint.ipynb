### Tensorflow, 텐서플로우
- 구글이 개발한 오픈소스 소프트웨어 라이브러리며, 머신러닝과 딥러닝을 쉽게 사용할 수 있도록 다양한 기능을 제공한다.
- 주로 이미지 인식이나 반복 신경망 구성, 기계 번역, 필기 숫자 반별 등을 위한 각종 신경망 학습에 사용된다.
- 딥러닝 모델을 만들 때, 기초부터 세세하게 작업해야 하기 때문에 진입장벽이 높다.

### Grayscale, RGB
- 흑백 이미지와 컬러 이미지는 각 2차원과 3차원으로 표현될 수 있다.
- 흑백 이미지는 0 ~ 255를 갖는 2차원 배열(높이 × 너비)이고,
  컬러 이미지는 0 ~ 255를 갖는 R, G, B 2차원 배열 3개를 갖는 3차원 배열 (높이 × 너비 × 채널)이다.

```python
from tensorflow.keras.datasets import fashion_mnist

# 데이터 세트 불러오기
# tuple 두개가 리턴됨
(train_images, train_targets), (test_images, test_targets) = fashion_mnist.load_data()
print("train datasets shape: ", train_images.shape, train_targets.shape)
print("test datasets shape: ", test_images.shape, test_targets.shape)
# 60000, 10000 = 배치 사이즈 = 배치 크기
# 지금 채널이 생략됨 ... 나중에 CNN할 때 뒤에 1로 해주기 지금은 채널 1 생략됨
```
